```python
# Функция для обучения и оценки модели
def evaluate_model(model, param_grid, X_train, y_train, X_val, y_val, model_name, param_name=None, param_scale='ordinary'):
    if model_name == 'LogisticRegression':
        # Используем LogisticRegressionCV для логистической регрессии
        model = LogisticRegressionCV(
            Cs=[100, 10, 1, 0.1, 0.01, 0.001],  # Значения C для подбора
            cv=5,  # Количество фолдов для кросс-валидации
            scoring='roc_auc',  # Метрика для оптимизации
            random_state=42,
            max_iter=1000,  # Увеличиваем количество итераций для сходимости
            refit=True  # Переобучение на всем датасете с лучшим C
        )
        # Обучение модели
        model.fit(X_train, y_train)

        # Лучшее значение C
        best_C = model.C_[0]
        print(f"Лучшее значение C для LogisticRegression: {best_C}")

        # Визуализация зависимости ROC-AUC от параметра C
        plt.figure(figsize=(10, 5))
        plt.title(f'{model_name}: C')
        plt.xscale('log')  # Логарифмическая шкала для C
        plt.plot(model.Cs_, model.scores_[1].mean(axis=0), label='mean ROC-AUC', color='red', lw=3)
        plt.fill_between(
            model.Cs_,
            model.scores_[1].mean(axis=0) - 2 * model.scores_[1].std(axis=0),
            model.scores_[1].mean(axis=0) + 2 * model.scores_[1].std(axis=0),
            color='green', label='filled area between errors', alpha=0.5
        )
        legend_box = plt.legend(framealpha=1).get_frame()
        legend_box.set_facecolor("white")
        legend_box.set_edgecolor("black")
        plt.xlabel('C')
        plt.ylabel('ROC-AUC')
        plt.show()

        return {'best_C': best_C}

    else:
        # Используем GridSearchCV для других моделей
        grid_search = GridSearchCV(
            model,
            param_grid,
            cv=5,
            scoring='roc_auc',
            n_jobs=-1,
            verbose=1
        )

        # Обучение модели
        grid_search.fit(X_train, y_train)

        # Лучшие параметры и оценка
        best_params = grid_search.best_params_
        best_score = grid_search.best_score_

        print(f"Лучшие параметры для {model_name}: {best_params}")
        print(f"Лучшее значение ROC-AUC (кросс-валидация): {best_score:.4f}")

        # Визуализация зависимости ROC-AUC от гиперпараметров
        if param_name:
            plt.figure(figsize=(10, 5))
            plt.title(f'{model_name}: {param_name}')

            if param_scale == 'log':
                plt.xscale('log')

            # Фильтруем результаты для выбранного гиперпараметра
            means = []
            errors = []
            grid = param_grid[param_name]

            for value in grid:
                # Фильтруем результаты для текущего значения гиперпараметра
                mask = [params[param_name] == value for params in grid_search.cv_results_['params']]
                means.append(np.mean(grid_search.cv_results_['mean_test_score'][mask]))
                errors.append(np.std(grid_search.cv_results_['mean_test_score'][mask]))

            plt.plot(grid, means, label='mean ROC-AUC', color='red', lw=3)
            plt.fill_between(
                grid,
                np.array(means) - 2 * np.array(errors),
                np.array(means) + 2 * np.array(errors),
                color='green', label='filled area between errors', alpha=0.5
            )
            legend_box = plt.legend(framealpha=1).get_frame()
            legend_box.set_facecolor("white")
            legend_box.set_edgecolor("black")
            plt.xlabel(param_name)
            plt.ylabel('ROC-AUC')
            plt.show()

        return {'best_params': best_params, 'best_score': best_score}
```

```python
# Модели и их параметры
models = {
    'LogisticRegression': {
        'model': make_pipeline(StandardScaler(), LogisticRegressionCV()),
        'param_grid': {},  # Параметры не нужны, так как используем LogisticRegressionCV
        'param_name': None,  # Для LogisticRegressionCV не нужен param_name
        'param_scale': 'ordinary'
    },
    'DecisionTreeClassifier': {
        'model': DecisionTreeClassifier(random_state=42),
        'param_grid': {
            'max_depth': [3, 5, 7, 10, 20],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        },
        'param_name': 'max_depth',  # Гиперпараметр для визуализации
        'param_scale': 'ordinary'
    },
    'RandomForestClassifier': {
        'model': RandomForestClassifier(random_state=42),
        'param_grid': {
            'n_estimators': [50, 100, 200],
            'max_depth': [3, 5, 7, 10, 20],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        },
        'param_name': 'n_estimators',  # Гиперпараметр для визуализации
        'param_scale': 'ordinary'
    }
}
```

```python
#Оценка всех моделей
results = {}
for model_name, model_info in models.items():
    print(f"Оценка модели: {model_name}")
    result = evaluate_model(
        model_info['model'],
        model_info['param_grid'],
        X_train, y_train,
        X_val, y_val,
        model_name,
        param_name=model_info.get('param_name'),  # Передаем param_name
        param_scale=model_info.get('param_scale', 'ordinary')  # Передаем param_scale
    )
    results[model_name] = result
    print("-" * 60)
```

```python
# Вывод лучших параметров и метрик
print("Итоговые результаты:")
for model_name, result in results.items():
    print(f"Модель: {model_name}")
    if model_name == 'LogisticRegression':
        print(f"Лучшее значение C: {result['best_C']}")
    else:
        print(f"Лучшие параметры: {result['best_params']}")
    print("-" * 60)
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import roc_auc_score


n_estimators_list = [50, 100, 200, 300]
learning_rates = [0.01, 0.05, 0.1, 0.2]

best_roc_auc = 0
best_params = {}

for n_estimators in n_estimators_list:
  for learning_rate in learning_rates:

    model = CatBoostClassifier(
            n_estimators=n_estimators,
            learning_rate=learning_rate,
            verbose=0
        )

    model.fit(X_train, y_train)

    y_pred_proba = model.predict_proba(X_val)[:, 1]

    roc_auc = roc_auc_score(y_val, y_pred_proba)

    if roc_auc > best_roc_auc:
        best_roc_auc = roc_auc
        best_params = {'n_estimators': n_estimators, 'learning_rate': learning_rate}

print(f"Лучшие параметры: {best_params}")
print(f"Лучший ROC-AUC: {best_roc_auc:.4f}")
```
