```python
X_train, X_val, y_train, y_val = train_test_split(data_encoded.values, y_data.values,
                                                    train_size=0.8,
                                                    random_state=42, stratify=y_data)
```

```python
def best_parameters(X_train, y_train, model, param_name, grid, param_scale='ordinary', draw=True):
    parameters = {param_name: grid}
    CV_model = GridSearchCV(estimator=model,
                            param_grid=parameters,
                            cv=5,
                            scoring='f1',
                            n_jobs=-1,
                            verbose=1)
    CV_model.fit(X_train, y_train)
    means = CV_model.cv_results_['mean_test_score']
    error = CV_model.cv_results_['std_test_score']

    if draw:
        plt.figure(figsize=(10, 5))
        plt.title(f'{model.__class__.__name__}: {param_name}')

        if param_scale == 'log':
            plt.xscale('log')

        plt.plot(grid, means, label='mean values of score', color='red', lw=3)
        plt.fill_between(grid, means - 2 * error, means + 2 * error,
                         color='green', label='filled area between errors', alpha=0.5)
        legend_box = plt.legend(framealpha=1).get_frame()
        legend_box.set_facecolor("white")
        legend_box.set_edgecolor("black")
        plt.xlabel('parameter')
        plt.ylabel('f1 score')
        plt.show()

    return CV_model.best_params_, CV_model.best_score_
```

```python
#    
models = [
    LogisticRegression(),  #  
    DecisionTreeClassifier(),  #  
    RandomForestClassifier()  #  
]
param_names = [
    'C',  #   LogisticRegression
    'max_depth',  #   DecisionTreeClassifier
    'n_estimators'  #   RandomForestClassifier
]
grids = [
    np.logspace(-3, 3, 10),  #   C ( )
    np.arange(1, 30),  #   max_depth
    np.arange(10, 200, 20)  #   n_estimators
]
param_scales = [
    'log',  #    C
    'ordinary',  #    max_depth
    'ordinary'  #    n_estimators
]
```

```python
best_results = {}

for model, param_name, grid, param_scale in zip(models,
                                                param_names,
                                                grids,
                                                param_scales):
    best_params, best_score = best_parameters(X_train, y_train, model, param_name, grid, param_scale)
    best_results[model.__class__.__name__] = {
        'best_params': best_params,
        'best_score': best_score
    }
```

```python
#     
for model_name, result in best_results.items():
    print(f": {model_name}")
    print(f" : {result['best_params']}")
    print(f"  f1: {result['best_score']:.4f}")
    print("-" * 40)
```

```python
train_auc = roc_auc_score(y_train, y_train_predicted)
test_auc = roc_auc_score(y_val, y_test_predicted)
```

```python
plt.figure(figsize=(10, 7))
plt.plot(*roc_curve(y_train, y_train_predicted)[:2], label='train AUC={:.4f}'.format(train_auc))
plt.plot(*roc_curve(y_val, y_test_predicted)[:2], label='test AUC={:.4f}'.format(test_auc))
legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
legend_box.set_facecolor("white")
legend_box.set_edgecolor("black")
plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
plt.show()
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import roc_auc_score


n_estimators_list = [50, 100, 200, 300]
learning_rates = [0.01, 0.05, 0.1, 0.2]

best_roc_auc = 0
best_params = {}

for n_estimators in n_estimators_list:
  for learning_rate in learning_rates:

    model = CatBoostClassifier(
            n_estimators=n_estimators,
            learning_rate=learning_rate,
            verbose=0
        )

    model.fit(X_train, y_train)

    y_pred_proba = model.predict_proba(X_val)[:, 1]

    roc_auc = roc_auc_score(y_val, y_pred_proba)

    if roc_auc > best_roc_auc:
        best_roc_auc = roc_auc
        best_params = {'n_estimators': n_estimators, 'learning_rate': learning_rate}

print(f" : {best_params}")
print(f" ROC-AUC: {best_roc_auc:.4f}")
```
