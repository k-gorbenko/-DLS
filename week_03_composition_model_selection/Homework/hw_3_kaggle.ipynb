# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
#поменяем тип некоторых данных
data['TotalSpent'] = data['TotalSpent'].str.strip()
data['TotalSpent'] = data['TotalSpent'].replace('', pd.NA)
data['TotalSpent'] = pd.to_numeric(data['TotalSpent'], errors='coerce')
data['TotalSpent'] = data['TotalSpent'].fillna(0)
data['TotalSpent'] = data['TotalSpent'].astype(float)
data['IsSeniorCitizen'] = data['IsSeniorCitizen'].astype(object)
```

```python
#разбиваю датасет на признаки и целевую переменную
X_data = data.drop('Churn', axis=1)
y_data = data['Churn']
```

```python
#Разбиваем признаки на качественные и количественные
numerical_columns = X_data.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = X_data.select_dtypes(include=['object']).columns
```

```python
plt.figure(figsize=(15, 10))
for i, column in enumerate(numerical_columns, 1):
  plt.subplot(2, len(numerical_columns), i)
  plt.hist(X_data[column], bins=30,  alpha=0.7)
  plt.title(f'Гистограмма {column}')

  plt.subplot(2, len(numerical_columns), i+len(numerical_columns))
  plt.boxplot(X_data[column])
  plt.title(f'Boxplot of {column}')

plt.tight_layout()
plt.show()
```

```python
#отрисовываем категориальные признаки
plt.figure(figsize=(50, 25))
for i, column in enumerate(categorical_columns, 1):
  value_counts = data[column].value_counts()

  plt.subplot(2, len(categorical_columns), i)
  value_counts.plot(kind='bar', color='green', alpha=0.7)
  plt.title(f'Bar plot of {column}')

  plt.subplot(2, len(categorical_columns), i+len(categorical_columns))
  plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=90)
  plt.title(f'Pie chart of {column}')

plt.tight_layout
plt.show()
```
