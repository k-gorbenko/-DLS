```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
encoded_data = pd.get_dummies(X_data, columns=[column for column in X_data.columns if X_data[column].dtype == 'object'])
X_data = X_data.drop([column for column in X_data.columns if X_data[column].dtype == 'object'], axis=1)
X_data = pd.concat([X_data, encoded_data], axis=1)
y_data = y_data.astype(int)
```

```python
scaler = StandardScaler()

columns_to_scale = ['ClientPeriod', 'MonthlySpending', 'TotalSpent']
X_data[columns_to_scale] = scaler.fit_transform(X_data[columns_to_scale])
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_data.values, y_data.values,
                                                    train_size=0.8,
                                                    random_state=42)
```

```python
def best_parameters(X_train, y_train, model, param_name, grid, param_scale='ordinary', draw=True):
    parameters = {param_name: grid}
    CV_model = GridSearchCV(estimator=model,
                            param_grid=parameters,
                            cv=5,
                            scoring='f1',
                            n_jobs=-1,
                            verbose=1)  #  verbose   
    CV_model.fit(X_train, y_train)
    means = CV_model.cv_results_['mean_test_score']
    error = CV_model.cv_results_['std_test_score']

    if draw:
        plt.figure(figsize=(15, 10))
        plt.title(f'{model.__class__.__name__}: {param_name}')

        if param_scale == 'log':
            plt.xscale('log')

        plt.plot(grid, means, label='mean values of score', color='red', lw=3)
        plt.fill_between(grid, means - 2 * error, means + 2 * error,
                         color='green', label='filled area between errors', alpha=0.5)
        legend_box = plt.legend(framealpha=1).get_frame()
        legend_box.set_facecolor("white")
        legend_box.set_edgecolor("black")
        plt.xlabel('parameter')
        plt.ylabel('f1 score')
        plt.show()

    return means, error
```

```python
#    
models = [
    LogisticRegression(),  #  
    DecisionTreeClassifier(),  #  
    RandomForestClassifier()  #  
]
param_names = [
    'C',  #   LogisticRegression
    'max_depth',  #   DecisionTreeClassifier
    'n_estimators'  #   RandomForestClassifier
]
grids = [
    np.logspace(-3, 3, 10),  #   C ( )
    np.arange(1, 30),  #   max_depth
    np.arange(10, 200, 20)  #   n_estimators
]
param_scales = [
    'log',  #    C
    'ordinary',  #    max_depth
    'ordinary'  #    n_estimators
]
```

```python
for model, param_name, grid, param_scale in zip(models,
                                                param_names,
                                                grids,
                                                param_scales):
    best_parameters(X_train, y_train, model, param_name, grid, param_scale)
```

```python
model = RandomForestClassifier(n_estimators=100, n_jobs=-1)

model.fit(X_train, y_train)
y_train_predicted = model.predict_proba(X_train)[:, 1]
y_test_predicted = model.predict_proba(X_val)[:, 1]
```

```python
train_auc = roc_auc_score(y_train, y_train_predicted)
test_auc = roc_auc_score(y_val, y_test_predicted)
```

```python
plt.figure(figsize=(10, 7))
plt.plot(*roc_curve(y_train, y_train_predicted)[:2], label='train AUC={:.4f}'.format(train_auc))
plt.plot(*roc_curve(y_val, y_test_predicted)[:2], label='test AUC={:.4f}'.format(test_auc))
legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
legend_box.set_facecolor("white")
legend_box.set_edgecolor("black")
plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
plt.show()
```
